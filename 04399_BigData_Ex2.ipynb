```python
import pandas as pd

data = pd.read_csv("/content/drive/MyDrive/04399_Big_Data_Analytics&Mining_Tech/data/JD_stock_data.csv", header=None, encoding='utf-8')

data.head()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

new_column_names = ['Date', 'open', 'high', 'low', 'close', 'adjclose', 'volume', 'ticker']
data.columns = new_column_names


data = data[1:].reset_index(drop=True)

data['Date'] = pd.to_datetime(data['Date'])

numeric_cols = ['open', 'high', 'low', 'close', 'adjclose', 'volume']
for col in numeric_cols:
    data[col] = pd.to_numeric(data[col], errors='coerce')

data.info()

plt.figure(figsize=(14, 7))
plt.plot(data['Date'], data['close'], label='Close Price')
plt.title('JD Stock Close Price Trend')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.grid(True)
plt.legend()
plt.show()
```
